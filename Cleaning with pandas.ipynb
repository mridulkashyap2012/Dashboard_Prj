# Manipulating the data

```python
df["Current Status"]
```

--------------> Making a new column "CF_Employee"

```python
def f(s):
    if s == 'Ex-Employees':
        return 0
    else:
        return 1
    
df["CF_Employee"]= df["Current Status"].apply(f)
df.CF_Employee
```

------------------> Filling up the "na" values in the coulumn "Education Domain"

```python
df["Education Domain"].fillna("other", inplace=True)
df["Education Domain"]
```

```python
df.Age
```

------------> Making a new column "CF_Age Band"

```python
def f(s):
    if s<25:
        a="Below 25"
        return a
    elif s<35 and s>=25:
        a="25-34"
        return a
    elif s<45 and s>=35:
        a="35-44"
        return a
    elif s<=55 and s>=45:
        a="45-54"
        return a
    elif s>55:
        a="Above 55"
        return a
    
df["CF_Age Band"]= df.Age.apply(f)
df["CF_Age Band"]
```

-------> Making a new column "M:1, F:0"

```python
df.Gender
```

```python
def f(s):
    if s=='Male':
        return 1
    else:
        return 0
    
df["M:1, F:0"]= df.Gender.apply(f)
df["M:1, F:0"]
```
